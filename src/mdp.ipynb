```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def gen_grid(w,h):
    return np.zeros((w,h))
```

```python
def show_grid(grid, rewards, pos):
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            if (i,j) == pos:
                plt.plot(i,j,'rx')
            else:
                plt.plot(i,j, 'bo')

            plt.text(i,j,str(rewards[i,j]))
    
```

```python
def action(grid, pos, act):
    actions = {"up": (0,1), "down": (0,-1), "right": (1,0), "left": (-1,0)}
    new_pos = tuple(x + y for x,y in zip(pos,actions[act]))
    if new_pos[0] < 0 or new_pos[0] >= grid.shape[0] or new_pos[1] < 0 or new_pos[1] >= grid.shape[1]:
        return pos
    return new_pos
```

```python
def get_reward(rewards, pos):
    return rewards[pos[0], pos[1]]
```

```python
def policy_evaluation(V, P, R, gamma=0.9, theta=1e-3):
    w, h = V.shape

    converged = False
    while not converged:
        delta = 0
        new_V = np.copy(V)
        new_P = np.copy(P)

        for i in range(w):
            for j in range(h):
                pos = (i,j)

                if pos == (3,3):
                    continue

                new_value = 0

                possible_actions = ["up", "down", "left", "right"]

                action_values = []
                for act in possible_actions:
                    next_pos = action(V, pos, act)
                    r = get_reward(R, next_pos)
                    v_next = V[next_pos[0], next_pos[1]]

                    action_values.append(r + gamma * v_next)

                best = np.argmax(action_values)
                best_value = action_values[best]
                new_value = best_value
                new_V[i,j] = best_value

                P[i,j] = best

                delta = max(delta, abs(new_value - V[i,j]))


        if delta < theta:
            converged = True

    return V, P
    
```

```python
values = gen_grid(4,4)
rewards = np.ones((4,4))*-1
rewards[3,3] = 100
policies = np.zeros((4,4))

cur_pos = (0,0)
possible_actions = ["up", "down", "left", "right"]

values, policies = policy_evaluation(values, policies, rewards)
print(policies)

plt.figure()
while cur_pos != (3,3):
    new_pos = action(values, cur_pos, possible_actions[int(policies[cur_pos[0], cur_pos[1]])])
    show_grid(values,values,cur_pos)
    plt.show()
    time.sleep(5)
```
